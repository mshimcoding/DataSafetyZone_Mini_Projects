## 오늘 날짜를 기준으로 인터파크에서 진행중인 뮤지컬 랭킹 데이터를 뽑아보세요 
- 항목 : 공연명, 장소, 기간, 점유율
- 주소 : http://ticket.interpark.com/contents/Ranking/RankList?pKind=01011&pCate=&pType=D&

```python
# import json
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
import pandas as pd
```

```python
url = 'http://ticket.interpark.com/contents/Ranking/RankList?pKind=01011&pCate=&pType=D'

# headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.88 Safari/537.36'}
# headers = {}
# request = Request(url, headers = headers)
request = Request(url)
print(request)
response = urlopen(request)
html = response.read()
soup = BeautifulSoup(html, 'html.parser')
soup
```

```python
soup.select('.rankingDetailBody > .rankBody > table > tr > td > .prdInfo > a > b')[0]
```

```python
soup.select('body > div.rankingDetailBody > div:nth-child(2) > table > tr > td.prds > div.prdInfo > a > b')

```

```python
soup.select('body > div.rankingDetailBody > div > table > tr > td.prds > div.prdInfo > a')
```

```python
name = soup.select('body > div.rankingDetailBody > div > table > tr > td.prds > div.prdInfo > a > b')
name
```

```python
name_l = []
for i in name :
    name_l.append(i.text)
```

```python
name_l
```

### 기간

```python
duration = soup.select('body > div.rankingDetailBody > div > table > tr > td.prdDuration')
```

```python
duration[0].text
```

```python
duration[0].text.replace('\r','').replace('\n','').replace('\t','')
```

```python
duration_l = []
for i in duration :
    t = i.text.replace('\r\n\r\n\r\n','').replace('\t                        ','')
    duration_l.append(t)
```

```python
duration_l
```

## 점유율

```python
percent = soup.select('body > div.rankingDetailBody > div > table > tr > td > b')
percent
```

```python
percent_l = []
for i in percent :
    percent_l.append(i.text)
```

```python
percent_l
```

#### 장소

```python
place = soup.select('body > div.rankingDetailBody > div > table > tr > td.prds > div.prdInfo > a ')
place[0].text.split('\t                                    ')[1].replace('\r\n\t','')

```

```python
# duration_l = []
# for i in duration :
#     t = i.text.replace('\r\n\r\n\r\n','').replace('\t                        ','')
#     duration_l.append(t)


place_l = []
for i in place :
    text = i.text.split('\t                                    ')[1].replace('\r\n\t                                ','')
    place_l.append(text)
    
place_l
```


### 전체 데이터 합치기

```python
temp = list(zip(name_l, place_l,duration_l, percent_l))
type(temp[0])
```

```python
data = pd.DataFrame(list(zip(name_l, place_l,duration_l, percent_l)), columns = ['Name', 'Place','Duration', 'Percent'])

```
