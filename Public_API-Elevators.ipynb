```python
url = "http://apis.data.go.kr/openapi/service/ElevatorInformationService/getElevatorList?"
serviceKey = "Wuei8mCxaANACez0WNnd0wigmv%2BKr3XEHWxKeKRS4AqA80j3vLXR7JARjfy78fIRYkKpHDN5ffyd7BkDBZR2nQ%3D%3D"
numOfRows = 10
pageNo = 1
sido = "%EC%84%9C%EC%9A%B8%ED%8A%B9%EB%B3%84%EC%8B%9C"
sigungu ="%EC%9D%80%ED%8F%89%EA%B5%AC"

api_url = url + 'serviceKey=' + serviceKey + '&numOfRows=' + str(numOfRows) + '&pageNo=' + str(pageNo) + '&sido=' + sido + '&sigungu=' + sigungu
req = requests.get(api_url)
html = req.text
soup = BeautifulSoup(html, 'html.parser')


add1 = []
add2 = []
bdNm = []
elNo = []
elSt = []
lasInsp = []
lLoad = []

t_add1 = soup.select('address1')
t_add2 = soup.select('address2')
t_bdNm = soup.select('buldnm')
t_elNo = soup.select('elevatorno') 
t_elSt = soup.select('elvtrstts')
t_lasInsp = soup.select('lastinspctde')
t_lLoad = soup.select('liveload')

#numOfRows도 ㅇㅋ 이지만, 마지막 페이지가 일부 비어서 나오니까 range() 는 len(add1)로 하자!!!

for i in range(len(t_add1)):
    add1.append(t_add1[i].text)
    add2.append(t_add2[i].text)
    bdNm.append(t_bdNm[i].text)
    elNo.append(t_elNo[i].text)
    elSt.append(t_elSt[i].text)
    lasInsp.append(t_lasInsp[i].text)
    lLoad.append(t_lLoad[i].text)

data = list(zip(add1, add2, bdNm, elNo, elSt, lasInsp, lLoad))
final_data = pd.DataFrame(data, columns=['주소1', '주소2', '건물명', '승강기번호', '승강기상태', '최종검사일', '적재하중' ])
final_data
```

```python
url = "http://apis.data.go.kr/openapi/service/ElevatorInformationService/getElevatorList?"
serviceKey = "Wuei8mCxaANACez0WNnd0wigmv%2BKr3XEHWxKeKRS4AqA80j3vLXR7JARjfy78fIRYkKpHDN5ffyd7BkDBZR2nQ%3D%3D"
numOfRows = 10
## pageNo = 1
sido = "%EC%84%9C%EC%9A%B8%ED%8A%B9%EB%B3%84%EC%8B%9C"
sigungu ="%EC%9D%80%ED%8F%89%EA%B5%AC"

add1 = []
add2 = []
bdNm = []
elNo = []
elSt = []
lasInsp = []
lLoad = []

callPages = 5
for i in range(callPages):
    api_url = url + 'serviceKey=' + serviceKey + '&numOfRows=' + str(numOfRows) + '&pageNo=' + str(i + 1) + '&sido=' + sido + '&sigungu=' + sigungu
    req = requests.get(api_url)
    html = req.text
    soup = BeautifulSoup(html, 'html.parser')

    t_add1 = soup.select('address1')
    t_add2 = soup.select('address2')
    t_bdNm = soup.select('buldnm')
    t_elNo = soup.select('elevatorno') 
    t_elSt = soup.select('elvtrstts')
    t_lasInsp = soup.select('lastinspctde')
    t_lLoad = soup.select('liveload')

    for j in range(len(t_add1)):
        add1.append(t_add1[j].text)
        add2.append(t_add2[j].text)
        bdNm.append(t_bdNm[j].text)
        elNo.append(t_elNo[j].text)
        elSt.append(t_elSt[j].text)
        lasInsp.append(t_lasInsp[j].text)
        lLoad.append(t_lLoad[j].text)

data = list(zip(add1, add2, bdNm, elNo, elSt, lasInsp, lLoad))
final_data = pd.DataFrame(data, columns=['주소1', '주소2', '건물명', '승강기번호', '승강기상태', '최종검사일', '적재하중' ])
final_data
```

```python
final_data.to_csv('elevatorInfo.csv', index=False, encoding='utf-8')
```
